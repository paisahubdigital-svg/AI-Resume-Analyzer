```python
from google.colab import files

uploaded = files.upload()

resume_path = list(uploaded.keys())[0]

print("Uploaded File:", resume_path)
```

```python
import fitz

def extract_text(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""

    for page in doc:
        text += page.get_text()

    return text

resume_text = extract_text(resume_path)

print(resume_text[:3000])
```

```python
skills_database = [

"Python",
"Java",
"C++",
"Machine Learning",
"Deep Learning",
"TensorFlow",
"PyTorch",
"SQL",
"MongoDB",
"Pandas",
"NumPy",
"Scikit-learn",
"Git",
"Docker",
"AWS",
"Azure",
"GCP",
"FastAPI",
"Flask",
"Django",
"Linux",
"Data Analysis",
"Power BI",
"Excel"

]

def extract_skills(text):

    found = []

    for skill in skills_database:

        if skill.lower() in text.lower():
            found.append(skill)

    return sorted(list(set(found)))

skills = extract_skills(resume_text)

print("Detected Skills:\n")

for s in skills:
    print("✅", s)
```

```python
total_skills = len(skills_database)
found = len(skills)

score = int((found / total_skills) * 100)

print("ATS Score:", score, "/100")
```

```python
missing = list(set(skills_database) - set(skills))

print("Missing Skills:\n")

for s in sorted(missing):
    print("❌", s)
```

```python
labels = ["Detected", "Missing"]
values = [len(skills), len(missing)]

plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.title("Resume Skill Analysis")
plt.show()
```
